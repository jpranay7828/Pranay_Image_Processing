# To read video using OPEN CV

```python
cap=cv2.VideoCapture(0);
while(True):
    ret, frame=cap.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame', gray)
    if cv2.waitKey(0) & ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
cap=cv2.VideoCapture(0);
while(cap.isOpened()):
    ret, frame = cap.read()
    cv2.imshow('frame', frame)
    if cv2.waitKey(0):
        break
cap.release()
cv2.destroyAllWindows()
```
